```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Leer el archivo CSV
data = pd.read_csv("C:\\Users\\Admin\\Desktop\\SoyHenry\\Proyectos\\Proyecto 2\\Archivos para trabajar\\ds_coursera")

# Extraer la columna deseada
columna_texto = data['name']

# Unir todos los textos en un solo string
texto_completo = ' '.join(columna_texto)

# Crear el objeto WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(texto_completo)

# Mostrar la nube de palabras utilizando matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
#Ultimo wordcloud con los 3 datasets juntos
# Lee los archivos CSV
data1 = pd.read_csv("C:\\Users\\Admin\\Desktop\\SoyHenry\\Proyectos\\Proyecto 2\\Archivos para trabajar\\ds_coursera")
data2 = pd.read_csv("C:\\Users\\Admin\\Desktop\\SoyHenry\\Proyectos\\Proyecto 2\\Archivos para trabajar\\ds3_edx")
data3 = pd.read_csv("C:\\Users\\Admin\\Desktop\\SoyHenry\\Proyectos\\Proyecto 2\\Archivos para trabajar\\ds4_udemy")

# Extrae la columna deseada (los títulos) de cada dataset
columna_texto1 = data1['name']
columna_texto2 = data2['title']
columna_texto3 = data3['course_title']
```

```python
# Combina todos los títulos en una sola cadena de texto
texto_completo = ' '.join(columna_texto1) + ' ' + ' '.join(columna_texto2) + ' ' + ' '.join(columna_texto3)
```

```python
# Crea el objeto WordCloud con las opciones de visualización deseadas
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(texto_completo)

# Muestra el Word Cloud utilizando matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
